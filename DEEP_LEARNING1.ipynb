```python
#Import Libraries
import tensorflow as tf
from tensorflow.keras import layers
import keras
from keras.datasets import mnist
import matplotlib.pyplot as plt 
```

```python
#Loading Dataset
(x_train_digit,y_train_digit),(x_test_digit,y_test_digit)=mnist.load_data()
```

```python
#Input training set
x_train_digit
```

```python
#Outout training set
y_train_digit
```

```python
#Names of number in the dataset in order
col_names=['Zero','One','Two','Three','Four','Five','Six','Seven','Eight','Nine'] 
```

```python
#Visualiazation the Digites
#Figure size width=10,Height=10
plt.figure(figsize=(10,10))
#Loops from 0 to 14(15 images) 
for i in range(15):
    #5,5 means 5 rows,5 col.i+1 means position
    plt.subplot(5,5,i+1)
    #Hide the xaxis num
    plt.xticks([])
    #Hide the yaxis num
    plt.yticks([])
    #display the ith digit image from training set,and image is gray sclae 
    plt.imshow(x_train_digit[i],cmap='gray')
    #Display the actual digit name below each image
    plt.xlabel(col_names[y_train_digit[i]])
#displays all the images together
plt.show()
```

```python
#Shows no of input test set,no of columns,no of rows
x_test_digit
```

```python
#shows no of output test set
y_test_digit.shape
```

```python
#input preprocessing
#ANN input image only in 2D Format .So we reshape the image
```

```python
x_train_digit=x_train_digit.reshape(60000,784)
x_test_digit=x_test_digit.reshape(10000,784)
```

```python
#Now the training input image is 2D format
x_train_digit.shape
```

```python
#Now the test input image is 2D format
x_test_digit.shape 
```

```python
#Output Preprocessing
#ANN output image only in categorical format
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
#The  output image is converted in to categorical format
y_train_digit=to_categorical(y_train_digit,num_classes=10)
y_test_digit=to_categorical(y_test_digit,num_classes=10)
```

```python
y_train_digit.shape
```

```python
y_test_digit.shape
```

```python
y_train_digit[5]
```

```python
y_test_digit[5]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#Create Base neural network
#sequential is used to create layer by layer neural network
#Dense shows the connectivity between neuron in each layer
models=keras.Sequential([layers.Dense(256,activation='relu',input_shape=(784,)),
                         layers.Dense(64,activation='relu'),        
                         layers.Dense(64,activation='relu'),        
                         layers.Dense(10,activation='sigmoid')])       
```

```python
#Input Layer Neuron 784
#1st Hidden Layer Neuron 256
#2nd Hidden Layer Neuron 64
#3rd Hidden Layer Neuron 64
#Output Layer Neuron 10
```

```python
#cerate a summmary for procedure(sequential)
models.summary()
```

```python
#how to calculate Dense parameter
#parameter=output_channel_number *(input_channel_number+1)
          # = 256*(784+1)
          #=256*785=>200960 
```

```python
#Compile the model
models.compile(loss="categorical_crossentropy",optimizer="adam",metrics=['accuracy'])
```

```python
#loss tells how the model perform wrongly.lowerloss=better model
#optimizer tells how to improve weights
#metrics calculate the accuracy of each epoch
```

```python
#work with fit
history=models.fit(x_train_digit,y_train_digit,batch_size=100,epochs=10,validation_data=(x_test_digit,y_test_digit))
```

```python
#total training sample/batch size=>60000/100=>600 steps per epoch
#1epoch=models sees all training data at once
```

```python
#predict the output of test data
y_predict=models.predict(x_test_digit)
```

```python
y_predict[0]
```

```python
#The above output is can't understand by user so we chande into normal value.that reason we use argmax.It shows the index of 
#maximum proability
y_predict=np.argmax(y_predict,axis=1)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Confusion matrix
#Using actual data and predict data we can find confusion matrix
con_matrix=confusion_matrix(y_test_digit_eval,y_predict)
#ggplot shows the output in grid background
plt.style.use('ggplot')
#set size of the figure
plt.figure(figsize=(10,10))
#show confusion matrix in heatmap.annot shows number inside each cell.annot_kws:size sets the font sizevof number in each cell
#linewidth adds thin line between cells.fmt format numbers as integers.
sns.heatmap(con_matrix,annot=True,annot_kws={'size':10},linewidths=0.5,fmt="d",cmap="gray")
#Add title to the plot
plt.title('True or Flase Predicted Digit MNIST\n',fontweight='bold',fontsize=15)
#display final output screen
plt.show()
```

```python
#In the above output rows is a actual digits.columns is a predicted digits.each cell value shows number of images.
#higher diagonal value is better model.ex (actual1 predict 1) is 1120
```

```python
from sklearn.metrics import classification_report
```

```python
#clssification report
#It compare actual and predicted labels.
print(classification_report(y_test_digit_eval,y_predict))
```

```python
#summarize the history for ACCURACY
#Model accuracy increase from low to high because because the model learns and updates its weights during each epoch. 
```

```python
#This plots training accuracy in red line
plt.plot(history.history['accuracy'])
#This plots test accuracy in blue line
plt.plot(history.history['val_accuracy'])
plt.title("Model_Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(['train','test'],loc='best')
plt.show()
```

```python
#summarize the history for LOSS
#Model loss decreases the from high to low because in each epoch the model minimizes the loss and reduces the error.
```

```python
#This plots training loss inred line
plt.plot(history.history['loss'])
#This plot test loss in blue line
plt.plot(history.history['val_loss'])
plt.title("Model_Loss")
plt.ylabel("Loss")
plt.xlabel("Epoch")
plt.legend(['train','test'],loc='best')
plt.show()
```
